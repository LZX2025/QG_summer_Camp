# 经典CNN架构对比学习笔记

## **一、LeNet-5 (1998)**
### 1. 核心贡献
- **首个成功的CNN**：手写数字识别(MNIST)
- **奠定CNN基础**：卷积层+池化层交替
- **激活函数**：Sigmoid/Tanh

### 2. 网络结构
```
INPUT(32×32) → CONV1(6@28×28) → POOL1(6@14×14) → 
CONV2(16@10×10) → POOL2(16@5×5) → FC1(120) → FC2(84) → OUTPUT(10)
```

### 3. 特点
- **小卷积核**：5×5
- **平均池化**：非重叠采样
- **参数量**：约60k


## **二、AlexNet (2012)**
### 1. 革命性突破
- **ImageNet冠军**：Top-5错误率15.3% → 8.0%
- **现代深度学习开端**
- **关键技术**：ReLU、Dropout、数据增强

### 2. 网络结构
```
INPUT(224×224) → [CONV1→POOL1] → [CONV2→POOL2] → 
[CONV3→CONV4→CONV5→POOL3] → FC1(4096) → FC2(4096) → OUTPUT(1000)
```

### 3. 创新点
- **ReLU激活**：解决梯度消失
- **双GPU训练**：分拆卷积层
- **LRN归一化**：局部响应归一化
- **参数量**：约60M


## **三、VGG (2014)**
### 1. 核心思想
- **深度堆叠小卷积核**：3×3替代大卷积核
- **结构统一**：仅通过卷积核数量区分不同版本

### 2. 网络结构(VGG16)
```
INPUT → [CONV3-64×2→POOL] → [CONV3-128×2→POOL] → 
[CONV3-256×3→POOL] → [CONV3-512×3→POOL] → [CONV3-512×3→POOL] → 
FC1(4096) → FC2(4096) → OUTPUT(1000)
```

### 3. 优势
- **更深网络**：16-19层
- **参数量**：约138M（VGG16）
- **迁移学习骨干网络**


## **四、ResNet (2015)**
### 1. 核心创新
- **残差块**：引入跳跃连接(Skip Connection)
- **解决退化问题**：训练更深网络(152层+)

### 2. 网络结构(ResNet18)
```
INPUT → CONV7-64→BN→ReLU→POOL → 
[RESBLOCK1(64)×2] → [RESBLOCK2(128)×2] → 
[RESBLOCK3(256)×2] → [RESBLOCK4(512)×2] → 
GAP → FC(1000)
```

### 3. 残差块结构
```
X → CONV→BN→ReLU→CONV→BN → +X → ReLU
  ↘____________________↗
```

### 4. 贡献
- **超深网络**：突破1000层
- **参数量**：ResNet18约11M
- **ImageNet Top-5错误率**：3.57%


## **五、架构对比表**
| 模型       | 层数 | 卷积核尺寸 | 激活函数 | 关键创新                 | 参数量   |
|------------|------|------------|----------|--------------------------|----------|
| LeNet-5    | 7    | 5×5        | Tanh     | 卷积+池化               | 60k      |
| AlexNet    | 8    | 11×11,5×5  | ReLU     | ReLU、Dropout            | 60M      |
| VGG16      | 16   | 3×3        | ReLU     | 小卷积核堆叠             | 138M     |
| ResNet18   | 18   | 3×3        | ReLU     | 残差连接                 | 11M      |


## **六、总结**
1. **LeNet**：CNN的起点，奠定基本框架
2. **AlexNet**：开启深度学习时代，引入ReLU和Dropout
3. **VGG**：证明深度和一致性的重要性
4. **ResNet**：通过残差连接突破深度瓶颈，现代架构基础
